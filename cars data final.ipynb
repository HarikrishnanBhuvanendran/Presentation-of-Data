# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
plt.figure(figsize=(8,8))
plt.hist(table["Car Gender"],color='teal',rwidth=1)
plt.title('graph of male vs female buyers participated in the sales',fontsize=12)
plt.xlabel('Car Gender')
plt.ylabel('frequency')
```
